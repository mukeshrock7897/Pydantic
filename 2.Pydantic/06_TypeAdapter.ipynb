### âœ… Basic Example

```python
from pydantic import TypeAdapter

adapter = TypeAdapter(list[str])
validated = adapter.validate_python(["fastapi", "pydantic"])
print(validated)  # ['fastapi', 'pydantic']
```

### ğŸ” Serialize with `dump_python()`
