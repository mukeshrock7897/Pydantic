#### ðŸ“˜ What is JSON Schema?

JSON Schema is a **standard way to describe JSON structure** â€” including:

* Required fields
* Data types
* Validation rules (like `minLength`, `maximum`, etc.)

Pydantic can auto-generate this from your models, which is **super useful** for:

* FastAPI docs
* API contract sharing
* Frontend form validation

---

### ðŸ§± Basic Usage

```python
from pydantic import BaseModel

class User(BaseModel):
    id: int
    name: str
    age: int | None = None

print(User.model_json_schema())
```

#### ðŸ” Output (formatted):

```json
{
  "title": "User",
  "type": "object",
  "properties": {
    "id": { "type": "integer" },
    "name": { "type": "string" },
    "age": { "type": "integer" }
  },
  "required": ["id", "name"]
}
```

---

### ðŸ§° Add Descriptions and Rules

```python
from pydantic import BaseModel, Field

class Product(BaseModel):
    name: str = Field(..., description="Name of the product")
    price: float = Field(..., gt=0, description="Price must be > 0")

print(Product.model_json_schema())
```
