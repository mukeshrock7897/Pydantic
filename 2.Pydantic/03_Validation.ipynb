## ‚úÖ **Validation ‚Äì Automatic & Custom**

### üõ°Ô∏è Automatic Validation

Pydantic automatically checks if the incoming data matches your type hints.

```python
from pydantic import BaseModel

class User(BaseModel):
    name: str
    age: int

# Correct input
user = User.model_validate({"name": "Mukesh", "age": 25})

# ‚ùå Incorrect input
# user = User.model_validate({"name": "Mukesh", "age": "twenty-five"})
```

The second one will raise a `ValidationError` because `age` must be an `int`.

---

### üß∞ Custom Validation (Function Style ‚Äì Pydantic v2)

You can write your own validation using `@field_validator`.

#### ‚úÖ Example:

```python
from pydantic import BaseModel, field_validator

class User(BaseModel):
    name: str
    age: int

    @field_validator("name")
    @classmethod
    def name_must_be_capitalized(cls, v):
        if not v.istitle():
            raise ValueError("Name must be capitalized")
        return v
```

Test it:

```python
User.model_validate({"name": "mukesh", "age": 30})  # ‚ùå will raise error
```

### ‚úÖ Multiple Field Validation (Advanced ‚Äì optional now)

```python
@model_validator(mode="after")
def check_all_fields(self):
    if self.age < 18 and "admin" in self.name.lower():
        raise ValueError("Too young to be admin")
    return self
```

We'll skip `model_validator` for now unless you want it ‚Äî focus on `field_validator` first.
