# 📎 FastAPI — File Uploads & Forms

### 🎯 Intent

Handle **forms** and **file uploads** in FastAPI with validation, limits, and streaming—ideal for dashboards and admin panels.

---

### 🧩 Core Components

1. **📝 Form Data (`Form`)**

   * Parse `application/x-www-form-urlencoded` or `multipart/form-data`.
   * Great for login forms, search, simple inputs.

2. **📂 File Uploads (`UploadFile` + `File`)**

   * `UploadFile` streams to temp file (memory efficient).
   * Access `.file`, `.filename`, `.content_type`.

3. **🧺 Multiple Files**

   * Accept `list[UploadFile]` for batch uploads.

4. **🧷 Mix Forms + Files**

   * Combine fields + file(s) in the same request (`multipart/form-data`).

5. **🛡️ Validation**

   * Check `content_type` (e.g., `image/png`).
   * Sanitize filenames before saving.

6. **💾 Safe Saving**

   * Save in chunks, never trust user-provided filenames.
   * Use secure paths or random names.

7. **📏 Size Limits**

   * Enforce upload limits via middleware/proxy.
   * Reject oversized files early.

8. **🧪 Testing**

   * Use `TestClient` with `files={...}`.
   * Assert correct handling of file name/type.

9. **🧰 Performance**

   * Prefer `UploadFile` (streaming) over `bytes` for large files.
   * Offload heavy work (resize, parsing) to BackgroundTasks/worker queues.

10. **🔒 Security**

* Never auto-execute uploaded files.
* Store outside web root.
* Strip metadata (EXIF/PII) before serving public files.
