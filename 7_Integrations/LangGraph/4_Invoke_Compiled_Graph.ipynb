# ▶️ LangGraph — Invoke Compiled Graph

### 🎯 Intent

Run a **compiled graph** with the Runnable API: `.invoke`, `.stream`, `.batch`.

---

### 🧩 Core Components

1. **⚙️ Compile**

   * `app = graph.compile()` → graph becomes a Runnable.

2. **▶️ `.invoke(state)`**

   * Run once with initial state.
   * Returns final merged state.

3. **🌊 `.stream(state)`**

   * Stream events/tokens live.
   * Useful for UIs & debugging.

4. **🗂️ `.batch([states])`**

   * Run multiple states in parallel.
   * Great for bulk evaluation.

5. **🧰 Config**

   * Add `tags`, `metadata`, `max_concurrency` per run.
   * Helps with tracing & throttling.

6. **📦 State Shape**

   * Input must follow the graph’s `State`.
   * Reducers merge updates during execution.

7. **🧪 Debugging**

   * Inspect `.stream()` outputs step by step.

8. **🔄 Iteration**

   * Feed final state back in for multi-turn agents.

9. **⚡ Performance**

   * Use `.batch()` + concurrency for throughput.
